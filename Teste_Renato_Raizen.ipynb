```python
import pandas as pd
import datetime as dt
from datetime import datetime
from pandas import Timestamp
import time
```

# *Obtendo as planilhas diretamente no site do governo*

Planilha de dados de combustíveis derivados de Petróleo, chamaremos ela de "dfCDP".

```python
dfCDP = pd.read_csv('https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/vdpb/vendas-derivados-petroleo-e-etanol/vendas-derivados-petroleo-etanol-m3-1990-2022.csv',sep=';' )
dfCDP.dtypes
```

Planilha de dados de combustíveis derivados de Óleo Diesel, chamaremos ela de "dfD".

```python
dfD = pd.read_csv('https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/vdpb/vct/vendas-oleo-diesel-tipo-m3-2013-2022.csv',sep=';')
dfD.dtypes
```

# *Efetuando o tratamento da planilha de combustíveis derivados de Petróleo.*

```python
dfCDP
```

```python
#Criando uma coluna 'created_at' do tipo timestamp
dfCDP['created_at'] = dt.datetime.now()
```

```python
#criando uma nova coluna chamada 'unit' do tipo string.
dfCDP['unit'] = '(m3)'
```

```python
#Trocando os valores dos meses para números, para implementar a junção das colunas para a criação de uma nova coluna.
dfCDP['MÊS'].replace(['JAN', 'FEV', 'MAR', 'ABR', 'MAI', 'JUN', 'JUL', 'AGO', 'SET', 'OUT', 'NOV', 'DEZ'],
                          ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12'], inplace=True)

#Renomeando as colunas
dfCDP.rename(columns={"UNIDADE DA FEDERAÇÃO": "uf", "PRODUTO": "product", "VENDAS": "volume" }, inplace=True)
```

```python
#Dropando a coluna 'GRANDE REGIÃO', pois não utilizaremos ele.
dfCDP.drop(['GRANDE REGIÃO'],axis=1,inplace=True)
```

```python
#Criando uma nova coluna chamada 'year_month' do tipo date, da junção das colunas 'ANO' e 'MÊS'.
dfCDP['year_month'] = dfCDP['ANO'].map(str) + '-' + dfCDP['MÊS'].map(str)
dfCDP.drop(['ANO', 'MÊS'],axis=1,inplace=True)
```

```python
#Organizando a tabela para melhor vizualização.
dfCDP = dfCDP[['year_month', 'uf', 'product', 'unit', 'volume', 'created_at']]
dfCDP.head()
```
